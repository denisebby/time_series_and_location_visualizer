```python
loc_df = pd.read_csv("../data/dunkin_stores.csv")
```

```python
loc_df.query("row_id == 1").state.iloc[0]
```

```python
df = loc_df.query("state == 'AL'")
```

```python
fig = go.Figure(data=go.Scattergeo(
        lon = df['loc_long'],
        lat = df['loc_lat'],
        text = df['row_id'],
        mode = 'markers',
        ))

fig.update_layout(
        title = 'Alabama stores',
        geo_scope='usa',
    )
fig.show()
```
